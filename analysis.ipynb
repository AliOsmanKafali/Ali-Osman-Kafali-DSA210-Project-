
## Spotify Tracks Data Analysis

This project analyzes a Spotify tracks dataset to explore the factors affecting song popularity. The analysis focuses on relationships between popularity and musical features such as energy, danceability, tempo, and explicit content.

The goal is to identify patterns and determine whether these features significantly influence how popular a song becomes.
> Add blockquote



```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/dataset.csv")
df.head()
```

```python
df = df.dropna()
```

```python
df["popularity"].hist()
plt.show()
```

Most songs have medium popularity.

```python
df[["popularity", "energy", "danceability", "tempo"]].corr()
```

The correlation analysis shows that popularity is not strongly related to energy, danceability, or tempo. Since all values are close to zero, these features do not significantly explain song popularity.

```python
plt.scatter(df["energy"], df["popularity"])
plt.xlabel("Energy")
plt.ylabel("Popularity")
plt.title("Energy vs Popularity")
plt.show()
```

The data points are densely and randomly distributed, suggesting no meaningful relationship between energy and popularity. This supports the earlier correlation results showing a very weak association.

```python
df["danceability"].hist()
plt.title("Danceability Distribution")
plt.show()
```

The distribution shows that most songs have medium danceability, with fewer songs having extremely low or high values.

```python
df_explicit = df[df["explicit"] == True]["popularity"]
df_non_explicit = df[df["explicit"] == False]["popularity"]

df_explicit.mean(), df_non_explicit.mean()
```

Explicit songs have a slightly higher average popularity compared to non-explicit songs. However, the difference is relatively small, indicating that explicit content is not a major factor in determining song popularity.

## Conclusion

The analysis shows that there is no strong relationship between popularity and most musical features such as energy, danceability, and tempo. Correlation values are very low, and scatter plots confirm weak or no clear patterns.

Additionally, while explicit songs have slightly higher average popularity, the difference is minimal. This suggests that popularity is likely influenced by other factors not included in this dataset, such as artist recognition, marketing, or external trends.
